#### Exercise 09. Is the average of "Valdemorillo" and "Galapagar" prices the same? (★★☆)

Print the both average prices and then write a conclusion about them

```python
import pandas as pd

df = pd.read_csv('assets/real_estate.csv', sep=';')

valdemorillo_prices = df[df['level5'] == 'Valdemorillo']['price']
galapagar_prices = df[df['level5'] == 'Galapagar']['price']

avg_price_valdemorillo = valdemorillo_prices.mean()
avg_price_galapagar = galapagar_prices.mean()


print("Precio promedio en Valdemorillo:", avg_price_valdemorillo)
print("Precio promedio en Galapagar:", avg_price_galapagar)


if avg_price_valdemorillo == avg_price_galapagar:
    print("Los precios promedio son iguales.")
else:
    print("Los precios promedio son diferentes.")
```

Después de calcular los promedios de precios para las poblaciones "Valdemorillo" y "Galapagar", podemos observar que los precios promedio no son iguales. Esto sugiere que hay una diferencia en los precios promedio entre estas dos poblaciones. Esta diferencia podría deberse a una variedad de factores, como la ubicación, las características de las propiedades y la demanda del mercado en cada área. 

#### Exercise 10. Is the average of "Valdemorillo" and "Galapagar" price per square meter (price/m2) the same? (★★☆)

Print the both average prices and then write a conclusion about

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
import pandas as pd

df = pd.read_csv('assets/real_estate.csv', sep=';')

valdemorillo_df = df[df['level5'] == 'Valdemorillo']
galapagar_df = df[df['level5'] == 'Galapagar']

valdemorillo_df['pps'] = valdemorillo_df['price'] / valdemorillo_df['surface']
galapagar_df['pps'] = galapagar_df['price'] / galapagar_df['surface']

avg_pps_valdemorillo = valdemorillo_df['pps'].mean()
avg_pps_galapagar = galapagar_df['pps'].mean()

print("Precio promedio por metro cuadrado en Valdemorillo:", avg_pps_valdemorillo)
print("Precio promedio por metro cuadrado en Galapagar:", avg_pps_galapagar)

if avg_pps_valdemorillo == avg_pps_galapagar:
    print("Los precios promedio por metro cuadrado son iguales.")
else:
    print("Los precios promedio por metro cuadrado son diferentes.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('assets/real_estate.csv', sep=';')

south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

south_belt_df = df[df['level5'].isin(south_belt_populations)]

median_prices = south_belt_df.groupby('level5')['price'].median()

median_prices.plot(kind='bar', color='skyblue')
plt.title('Mediana de los precios en el "cinturón sur" de Madrid')
plt.xlabel('Población')
plt.ylabel('Mediana del precio')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

#### Exercise 16. Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (you must use the subset obtained in the question 14) (★★★)

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('assets/real_estate.csv', sep=';')

south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

south_belt_df = df[df['level5'].isin(south_belt_populations)]

normalized_prices = {}
for population in south_belt_populations:
    prices = south_belt_df[south_belt_df['level5'] == population]['price']
    normalized_prices[population] = (prices - prices.min()) / (prices.max() - prices.min())

plt.figure(figsize=(10, 6))
for population in south_belt_populations:
    sns.histplot(normalized_prices[population], label=population, alpha=0.5, kde=True)

plt.title('Histograma normalizado de precios por población en el "cinturón sur" de Madrid')
plt.xlabel('Precio normalizado')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('assets/real_estate.csv', sep=';')

populations_of_interest = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

subset = df[df['level5'].isin(populations_of_interest)]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))


for i, (population, ax) in enumerate(zip(populations_of_interest, axes.flatten())):
    population_subset = subset[subset['level5'] == population]
    ax.scatter(population_subset['surface'], population_subset['price'], label=population, alpha=0.5)
    ax.set_title(f'Scatter plot for {population}')
    ax.set_xlabel('Surface')
    ax.set_ylabel('Price')
    ax.grid(True)
    ax.legend()

plt.tight_layout()
plt.show()
```
